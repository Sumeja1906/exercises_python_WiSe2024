```python
import pandas as pd 
import plotly.express as px 
import numpy as np
import seaborn as sns
```

```python
df_final = pd.read_csv("data/stocks.csv", 
                       sep = ",", 
                       decimal = ".").drop(['Unnamed: 0'], axis=1)

df_final.dropna(inplace = True)
```

```python
df_final.head()
```

```python
df_graph_long = pd.melt(
    df_final,
    id_vars = 'Date',
    var_name = 'stock',
    value_name = 'value'
)

df_graph_long
```

```python
px.line(df_final, 
        x = 'Date', 
        y = ['AAPL', 'GOOG', 'META', 'MSFT'],
        labels = {"value": "Dollar",
                  "Date": "Zeit",
                  "variable": "Aktie"},
        title = 'Schlusskurs')
```

```python
df_pc = df_final.set_index('Date').pct_change().reset_index()
```

```python
import numpy as np
sns.heatmap(df_pc.select_dtypes(include = np.number).corr(), annot = True);
```

```python
df_pc_long = pd.melt(
    df_pc,
    id_vars = 'Date',
    var_name = 'stock',
    value_name = 'value'
)

px.line(df_pc_long,
        x = 'Date', 
        y = 'value', 
        color = "stock",
        labels = {"value": "Rendite",
                  "Date": "Zeit",
                  "stock": "Aktie"},
        title = 'Aktienrenditen')
```

```python
df_pc.select_dtypes(include = np.number).describe()
```

```python
px.histogram(df_pc.select_dtypes(include = np.number),
            labels = {"value": "Wert",
                      "variable": "Aktie"},
            title = 'Verteilung der Aktienrenditen')
```

```python
df_cret = df_pc.dropna()

df_cumulative_returns = df_cret.set_index('Date').apply(
    lambda x: (1 + x).cumprod() - 1).reset_index()

df_cumulative_returns
```

```python
px.line(df_cumulative_returns, 
        x = 'Date', 
        y = ['AAPL', 'GOOG', 'META', 'MSFT'],
        labels = {'value': 'Kumulierte Rendite',
                  'Date': 'Zeit',
                  'variable': 'Stock'},
        title = 'Kumulierte Rendite')
```
